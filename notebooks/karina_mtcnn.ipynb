```python
from cv2 import imread
from mtcnn.mtcnn import MTCNN
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
```

```python
img = '../raw_data/train/fear/img_95.png'
```

```python
pixels_img = imread(img)
```

```python
detector = MTCNN()
result = detector.detect_faces(pixels_img)
```

```python
result[0]
```

```python
def draw_image_with_boxes(img, result_list, with_dots=False):
    
    data = plt.imread(img)
    plt.imshow(data)

    ax = plt.gca()
    for result in result_list:
        # get coordinates
        x, y, width, height = result['box']
        # create the shape
        rect = Rectangle((x, y), width, height, fill=False, color='red')
        # draw the box
        ax.add_patch(rect)
        if with_dots:
            for key, value in result['keypoints'].items():
                # create and draw dot
                dot = Circle(value, radius=2, color='red')
                ax.add_patch(dot)
    plt.show()
```

```python
draw_image_with_boxes(img, result, with_dots=True)
```

```python
def draw_faces(img, result_list):
    
    data = plt.imread(img)
    for i in range(len(result_list)):
        # get coordinates
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height
        # define subplot
        plt.subplot(1, len(result_list), i+1)
        plt.axis('off')
        plt.imshow(data[y1:y2, x1:x2])
    plt.show()
```

```python
draw_faces(img, result)
```

```python
img_family = '../raw_data/family.jpg'
pixels_family = plt.imread(img_family)
results = detector.detect_faces(pixels_family)
```
